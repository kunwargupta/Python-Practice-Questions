```python
sales = [1200, 3400, 1500, 800, 2100, 3000, 2700, 900]
```

```python
res = [x for x in sales if x > 2000]
res
```

```python
# 1. Write a program to print whether each sale is above or below 2000.

for i in sales:
    if i > 2000:
        print(f'{i}: Above')
    else:
        print(f'{i}: Below')
```

```python
# 2. Count how many sales are above 2500 using a for loop.
count = 0
for i in sales: 
    if i > 2500:
        count += 1
count
```

```python
# 3. Use if-elif-else to classify each sale as “Low (<1500)”, “Medium (1500–2500)”, or “High (>2500)”.

for i in sales:
    if i < 1500:
        print(f'{i}: Low')
    elif (i > 1500) and (i < 2500):
        print(f'{i}: Medium')
    else:
        print(f'{i}: High')
```

```python
# 4. Using a while loop, print total sales until the cumulative sum exceeds 10,000.
sales = [1200, 3400, 1500, 800, 2100, 3000, 2700, 900]
cum_sum = 0
i = 0
while i < len(sales) and cum_sum <= 10000:
    cum_sum += sales[i]
    i += 1

cum_sum
```

```python
# 5. Use break to stop iteration once a sale greater than 3000 is found

i = 0
while i < len(sales):
    if sales[i] > 3000:
        break
    print(sales[i])
    i += 1
```

```python
# 6. Use continue to skip sales below 1000.
sales = [1200, 3400, 1500, 800, 2100, 3000, 2700, 900]
i = 0
while i < len(sales):
    if sales[i] <  1000:
        i += 1
        continue #from here it'll again go to while and skip below code and checks again i < len(sales) for that put i += 1 before to prevent infinite loop 
    print(sales[i])
    i += 1
```

```python
# 7. Add an else clause in a loop to print “Processed all sales”.
sales = [1200, 3400, 1500, 800, 2100, 3000, 2700, 900]
for i in sales:
    print(i)
else:
    print("Processed all sales")
```

```python
# 8. Calculate the average sale without using sum() or len().

total = 0
count = 0
for i in sales:
    total += i
    count += 1

print(total//count) #integer division
```
